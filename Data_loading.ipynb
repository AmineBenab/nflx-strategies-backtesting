```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ticker = "SPY"
data = yf.download("SPY",start = "2018-01-01",end = "2025-01-01",interval = "1d", auto_adjust = True)
data = data.dropna()
data_return = np.log(data["Close"]/data["Close"].shift(1))
data_return = data_return.dropna()

ticker = "NFLX"
data = yf.download("NFLX",start = "2018-01-01",end = "2025-01-01",interval = "1d", auto_adjust= True)
data = data.dropna()
data_return = np.log(data["Close"]/data["Close"].shift(1))
data_return = data_return.dropna()
data.to_csv("NFLX_data.csv")
data_return.to_csv("NFLX_data_return.csv")
```
