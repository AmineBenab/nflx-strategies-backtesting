```python
#Stratégie de Buy and Hold
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skew, kurtosis
data = pd.read_csv("SPY_data.csv",index_col=0, parse_dates = True)
data_return = pd.read_csv("SPY_data_return.csv",index_col=0, parse_dates = True).squeeze()
V0 = 100
cum_return = np.exp(data_return.sum())-1
mu = data_return.mean()
sigma = data_return.std()
skewness = skew(data_return)
kurto = kurtosis(data_return)
T = len(data_return)
anu_return = (1+cum_return)**(252/T)-1
anu_vol = sigma*np.sqrt(252)
sharpe = anu_return / anu_vol
winrate= (data_return>0).sum()/ T
best_day = data_return.max()
worst_day = data_return.min()

metrics = ["Cumulative Return","Annualized Return","Mean","Volatility", "Annualized Volatility", "Sharpe Ratio", "Skewness", "Kurtosis","Winrate","Best Day", "Worst Day"]
value = [f"{cum_return:.2%}",f"{anu_return:.2%}",f"{mu:.2%}",f"{sigma:.2%}",f"{anu_vol:.2%}",
        f"{sharpe:.4f}",f"{skewness:.4f}",f"{kurto:.4f}",
        f"{winrate:.2%}",f"{best_day:.2%}",f"{worst_day:.2%}"]
df_metrics = pd.DataFrame({
    "Metrics": metrics,
    "Values" : value
})
x = np.linspace(data_return.min(),data_return.max(),len(data_return))
plt.hist(data_return, bins=100, density = True, alpha = 0.7)
plt.plot(x,norm.pdf(x,mu,sigma), color = "green", linewidth = 2, label = f"N({mu:.4f},{sigma:.4f})")
plt.title("Log-returns Histogram")
plt.xlabel("Log Return")
plt.grid(True)
plt.show()

data_return.plot(figsize =(12,5), title = "Daily Return SPY")
plt.ylabel("Log Return")
plt.xlabel("Date")
plt.grid(True)
plt.show()

equity_gain = V0*(1 + data_return).cumprod()
plt.plot(equity_gain, color = "orange", linewidth = 2)
plt.title("Equity Curve of SPY")
plt.ylabel("Portfolio Value")
plt.grid(True)
plt.show()

pic = equity_gain.cummax()
drawdown = (equity_gain - pic)/ pic
plt.plot(drawdown, color = "red", linewidth = 2)
plt.title(" Drawdown Curve")
plt.ylabel("Drawdown")
plt.grid(True)
plt.show()

roll = data_return.rolling(21).std().dropna() * np.sqrt(252)
plt.plot(roll, color = "pink", linewidth = 2)
plt.title("Rolling Volatility")
plt.ylabel("Annualized Volatility")
plt.grid(True)
plt.show()

df_metrics
```
